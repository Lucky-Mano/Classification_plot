```python
from keras.datasets import mnist
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [12, 9]
```

```python
(x_train, x_test), (y_train, y_test) = mnist.load_data()
```

```python
train = np.concatenate((x_train, y_train))
test = np.concatenate((x_test, y_test))
```

# PCA

```python
pca = PCA(n_components=2)
```

```python
%%time
reduced = pca.fit_transform(train.reshape(-1, 28 * 28))
```

```python
for i in range(10):
    num = reduced[test == i]
    plt.scatter(num[:, 0], num[:, 1], s=5, label=f'{i}')

plt.legend()
plt.show()
```

# t-SNE

```python
tsne = TSNE(learning_rate=100.0)
```

```python
%%time
reduced = tsne.fit_transform(train.reshape(-1, 28 * 28))
```

```python
for i in range(10):
    num = reduced[test == i]
    plt.scatter(num[:, 0], num[:, 1], s=5, label=f'{i}')

plt.legend()
plt.show()
```
